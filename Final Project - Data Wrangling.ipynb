Dataset link :

Perform following analysis on above dataset :

1. Import the dataset using Pandas from above mentioned url.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(Url)
df.head()
```

2. High Level Data Understanding:
    a. Find no. of rows & columns in the dataset.
    b. Data types of columns.
    c. Info & describe of data in dataframe.

```python
# a. Find no. of rows & columns in the dataset.
num_rows, num_cols = df.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
#b. Data types of columns.
```

```python
print("\nData types of columns:")
print(df.dtypes)
```

```python
# c. Info & describe of data in dataframe.
print("\nData Info:")
print(df.info())
print("\nData Description:")
print(df.describe())
```

3. Low Level Data Understanding :
 a. Find count of unique values in location column.
 b. Find which continent has maximum frequency using values counts.
 c. Find maximum & mean value in 'total_cases'.
 d. Find 25%,50% & 75% quartile value in 'total_deaths'.
 e. Find which continent has maximum 'human_development_index'.
 f. Find which continent has minimum 'gdp_per_capita'.

```python
# a. Find count of unique values in location column.
print("\nCount of unique values in location column:")
print(df['location'].nunique())
```

```python
# b. Find which continent has maximum frequency using values counts.
print("\nContinent with maximum frequency:")
print(df['continent'].value_counts().idxmax())
```

```python
# c. Find maximum & mean value in 'total_cases'.
print("\nMaximum value in 'total_cases':", df['total_cases'].max())
print("Mean value in 'total_cases':", df['total_cases'].mean())
```

```python
# d. Find 25%,50% & 75% quartile value in 'total_deaths'.
print("\n25% quartile value in 'total_deaths':", df['total_deaths'].quantile(0.25))
print("50% quartile value in 'total_deaths':", df['total_deaths'].quantile(0.50))
print("75% quartile value in 'total_deaths':", df['total_deaths'].quantile(0.75))
```

```python
# e. Find which continent has maximum 'human_development_index'.
max_hdi_continent = df.loc[df['human_development_index'].idxmax(), 'continent']
print("\nContinent with maximum 'human_development_index':", max_hdi_continent)
```

```python
# f. Find which continent has minimum 'gdp_per_capita'.
min_gdp_continent = df.loc[df['gdp_per_capita'].idxmin(), 'continent']
print("\nContinent with minimum 'gdp_per_capita':", min_gdp_continent)
```

4. Filter the dataframe with only this columns ['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index'] and update the data frame.

```python
selected_columns = ['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index']
df = df[selected_columns]
df.head()
```

5. Data Cleaning:
 a. Remove all duplicates observations.
 b. Find missing values in all columns.
 c. Remove all observations where continent column value is missing.
 Tip : using subset parameter in dropna
 d. Fill all missing values with 0.

```python
# a. Remove all duplicates observations.
df.drop_duplicates(inplace=True)
print(df.head())
```

```python
# b. Find missing values in all columns.
print("\nMissing values in all columns:")
print(df.isnull().sum())
print(df.head())
```

```python
# c. Remove all observations where continent column value is missing.
df.dropna(subset=['continent'], inplace=True)
print(df.head())
```

```python
# d. Fill all missing values with 0.
df.fillna(0, inplace=True)
print(df.head())
```

6. Date time format :
 a. Convert date column in datetime format using pandas.to_datetime.
 b. Create new column month after extracting month data from date column.

```python
# a. Convert date column in datetime format using pandas.to_datetime.
df['date'] = pd.to_datetime(df['date'])
print(df.head())
```

```python
# b. Create new column month after extracting month data from date column.
df['month'] = df['date'].dt.month
print(df.head())
```

7. Data Aggregation:
 a. Find max value in all columns using groupby function on 'continent'column.
 Tip: use reset_index() after applying groupby
 b. Store the result in a new dataframe named 'df_groupby'.
 (Use df_groupby dataframe for all further analysis)

```python
# a. Find max value in all columns using groupby function on 'continent'column.
# b. Store the result in a new dataframe named 'df_groupby'.
df_groupby = df.groupby('continent').max().reset_index()
print(df_groupby.head())
```

8. Feature Engineering :
 a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to    'total_cases'.

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
df_groupby.head()
```

9. Data Visualization :
 a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist     plot.
 b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'.
 c. Plot Pairplot on df_groupby dataset.
 d. Plot a bar plot of 'continent' column with 'total_cases' .
 Tip : using kind='bar' in seaborn catplot.

```python
# a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist     plot.
sns.distplot(df['gdp_per_capita'])
plt.title("Histogram of GDP per Capita")
plt.xlabel("GDP per Capita")
plt.show()
```

```python
# b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'.
sns.scatterplot(data = df, x='total_cases', y='gdp_per_capita')
plt.title("Scatter Plot of Total Cases vs GDP per Capita")
plt.show()
```

```python
# c. Plot Pairplot on df_groupby dataset.
sns.pairplot(df_groupby)
plt.title("Pair Plot of Grouped Data")
plt.show()
```

```python
# d. Plot a bar plot of 'continent' column with 'total_cases' .
sns.catplot(data=df_groupby, x='continent', y='total_cases', kind='bar')
plt.title("Total Cases by Continent")
plt.xticks(rotation=45)
plt.show()
```
